## NIH does not release the gender of the scientists it funds

First a little bit about how I got the NIH data. NIH releases a huge amount of data about funded grants. Individual grants can be searched for by entering search terms in the NIH RePORTER website. Data for each fiscal year can be downloaded from the ExPORTER. Downloading multiple files was made super easy by using this awesome download manager tool/plugin for Mozilla called not shockingly DownThemAll. 

The grant data is extremely rich. For every funded grant there is a information that includes the funding institute, the Principal Investigatoe (PI) name, university, type of grant, amount key words etc. However there were two important pieces of  information that are not released. First there is no publicly available information about rejected grants, or how many rounds of review a  particular grant went through before it was funded. Second no identifying information about the PI such as gender, ethnicity, age etc. is released

## Generating a gender index

My initial approach was to take a pattern classification approach using some obvious heuristics mostly to do with the last letter of each scientists first name. For example female names usually end with the letter ‘I’ or ‘A’ . Male names often end with a ‘T’ or ‘N’. However this heuristic fails for all of the ambiguous letters . Turns out though this approach has already been implemented in Python using the NLTK package, but at best has an accuracy of ~81%.  This wasn’t quite good enough, because an error rate of 20% could seriously skew our interpretation of gender parity in NIH funding. 

So I took an alternate approach. I cross-referenced the scientist names with the census database. So after some frenzied coding in Python, I generated a gender index for each PI ranging for -1 to +1. I'll walk you through my code below.

```python
import os
import re
import csv

import pandas as pd
from pandas import DataFrame,Series
import numpy as np
import random
```

```python
def getGenderIndex(PIfirstNames):
    # provide a list of names all in lowercase
    groupedNames=pd.read_csv('/datascience/nihProject/rawData/names1950onwards.csv') # read in a consolidated name and gender frequency file
    groupedNames.name[groupedNames.name.isnull()]='noName' # account for missing data
    groupedNames.name=[name.lower() for name in groupedNames.name]
    groupedNames=groupedNames.set_index('name')

    genderIndex=pd.DataFrame()
    

    for name in PIfirstNames:
        try:
            repsByGender=groupedNames.ix[name]
            if repsByGender.shape[0] ==3: # if there is a match corresponding to a single gender find out if male or female
                if 'M' in repsByGender['gender']: ind= -1 # check for up dict keys, 
                elif 'F' in repsByGender['gender']: ind=1
                genderIndex=np.append(genderIndex,ind)
            
            elif repsByGender.shape[0] == 2 : # if match for both gender generate gender index
                # this index estimates how much the gender of a particular name is contaminated by the other gender. 
                # the sign of the gender index is also automatically determined by the more common gender
                ind=(repsByGender.reps[0]-repsByGender.reps[1])/(repsByGender.reps[0]+repsByGender.reps[1])
                genderIndex=np.append(genderIndex,ind)

        except KeyError:
            genderIndex=np.append(genderIndex,np.nan)
        
        
        
    genderIndex.reshape(-1,1)
    PIgender=pd.DataFrame({'firstName':PIfirstNames,'gender':genderIndex})

    return PIgender
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import mpld3

# set grant-type and load in the data. This will later be done iteratively to visualize the data for all the different grant types
grantType='R01'
numGrants=pd.read_csv('/datascience/nihProject/rawData/%snumByGender2014.csv'%grantType,index_col=0)
costGrants=pd.read_csv('/datascience/nihProject/rawData/%scostByGender2014.csv'%grantType,index_col=0)

costGrants['sum']=np.divide(costGrants['sum'],10000) # generate the total amount awarded in billions of dollars
years = numGrants['year'].unique()
```

I also wrote a quick and dirty linear regression function so we can look at trends in the data

```python
import statsmodels.formula.api as sm
from sklearn.linear_model import LinearRegression
import scipy, scipy.stats
# --------------------------------------------------------------------------------
# --------------------------------------------------------------------------------
def doLinearReg(X,Y):
    # enter a data frame that has atleast one X and one Y column
    
    intercept = np.ones((len(X),1))
    #Y = data.iloc[:,2]
    X = np.column_stack((X,intercept))
    result = sm.OLS( Y, X ).fit()
    params=result.params
    return params
```

```python
clrs=['#1f77b4','#ff7f0e','#2ca02c', '#d62728']
grantType=['F31','F32','K99','R01']
plt.rcParams['figure.figsize'] = 7,5
fig, ax3= plt.subplots(1)
label=list()
labels=list()
yearEqual=list()
coll=list()
for i in range(0,4): 
    numGrants=pd.read_csv('/datascience/nihProject/rawData/%snumByGender2014.csv'%grantType[i],index_col=0)
    years = numGrants['year'].unique().tolist()

    ratio=np.true_divide(numGrants['percent'].ix[0],numGrants['percent'].ix[2]).tolist()

    paramsRatio=doLinearReg(years,ratio)
    fittedRatio=np.multiply(paramsRatio[0],years)+ paramsRatio[1]
    yearEqual.append((1-paramsRatio[1])/paramsRatio[0])

    fig1=ax3.scatter(years,ratio,s=30,color=clrs[i]) 
    line1=ax3.plot(years,fittedRatio, color=clrs[i], linestyle='-', linewidth=2)
    coll.append(line1)
    
    labels= ['Ratio M:W = {0}'.format(j) for j in np.round(ratio,decimals=1)]
    label.append(labels)
    tooltip = mpld3.plugins.PointLabelTooltip(fig1, labels=labels)
    mpld3.plugins.connect(fig, tooltip)
#ax.set_title('Dollar value of RO1s by gender')
ax3.set_ylabel(' Grants awarded to men:women')
ax3.set_xlabel('fiscal year')
ax3.set_ylim(0,4)
ax3.set_xlim((2000,2016))
ax3.set_xticks(range(2000,2016,4))

line_collections = [coll[0],coll[1],coll[2] ,coll[3]]
interactive_legend = plugins.InteractiveLegendPlugin(line_collections, grantType)
plugins.connect(fig, interactive_legend)

plt.subplots_adjust(right=.7)

mpld3.enable_notebook()
mpld3.display()
```
